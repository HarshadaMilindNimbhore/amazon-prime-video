# Rating wise Genre distrbition 

```python
# Split the 'Favorite Genres' column
df_split = df.assign(Favorite_Genres=df['Favorite Genres'].str.split(',')).explode('Favorite_Genres')

# Calculate the average rating for each genre
average_ratings = df_split.groupby('Favorite_Genres')['Feedback/Ratings'].mean().reset_index()
average_ratings.columns = ['Favorite_Genres', 'Average_Rating']

# Set the color palette
colors = sns.color_palette('pastel')

# Plot the pie chart
plt.figure(figsize=(10, 7))
plt.pie(average_ratings['Average_Rating'], labels=average_ratings['Favorite_Genres'], autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Average Rating for Each Favorite Genre')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# from the above pie chart we see that Comedy and Drama type movies and videos having high rating as compair to other genre 
```

# Age group wise generes distribution and Gender wise generes distribution

```python
# Define a function to categorize ages
def categorize_age(age):
    if age < 18:
        return 'Child'
    elif 18 <= age < 30:
        return 'Adult'
    elif 30 <= age < 60:
        return 'Teenager'
    else:
        return 'Senior Citizen'

# Apply the function to create the 'Age_Group' column
df['Age_Group'] = df['Age'].apply(categorize_age)

# Split the 'Favorite Genres' column
df_split = df.assign(Favorite_Genres=df['Favorite Genres'].str.split(',')).explode('Favorite_Genres')

# Group by age group and favorite genres, and count the number of people in each group
age_genre_counts = df_split.groupby(['Age_Group', 'Favorite_Genres']).size().reset_index(name='Count')

# Plot the data
plt.figure(figsize=(14, 7))
sns.barplot(data=age_genre_counts, x='Age_Group', y='Count', hue='Favorite_Genres', palette='pastel')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Count of People by Age Group and Favorite Genre')
plt.legend(title='Favorite Genres', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# most of the adult waching Horror Movies 
# most of the Senior Citizen Action Movies 
# Most of the teenager watching Documentry type Movies 
```

```python
# Split the 'Favorite Genres' column
df_split = df.assign(Favorite_Genres=df['Favorite Genres'].str.split(',')).explode('Favorite_Genres')

# Group by gender and favorite genres, and count the number of people in each group
gender_genre_counts = df_split.groupby(['Gender', 'Favorite_Genres']).size().reset_index(name='Count')

# Plot the data
plt.figure(figsize=(12, 8))
sns.barplot(data=gender_genre_counts, x='Gender', y='Count', hue='Favorite_Genres', palette='pastel')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of People by Gender and Favorite Genre')
plt.legend(title='Favorite Genres', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# most of the females waching action Movies 
# most of the males waching Horror Movies 
```

# Gender wise Device used 

```python
# Group by gender and devices used, and count the number of people in each group
gender_device_counts = df.groupby(['Gender', 'Devices Used']).size().reset_index(name='Count')

# Plot the data
plt.figure(figsize=(12, 8))
sns.barplot(data=gender_device_counts, x='Gender', y='Count', hue='Devices Used', palette='pastel')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of People by Gender and Devices Used')
plt.legend(title='Devices Used', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# from the above graph most of the male and females are used smartphones for the waching amazon prime video 
```
